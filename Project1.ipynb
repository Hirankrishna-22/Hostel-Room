5. Hostel Room Allocation System
Build a Hostel Room Allocation and Management System.
The program should manage rooms with defined capacities and student registrations.
Students can be allocated rooms based on availability and constraints such as year or
gender (optional).
The system must ensure one student occupies only one room at a time. Vacating rooms,
viewing occupancy status, and generating allocation reports should be supported

```python
class Student:
    def __init__(self, sid, name, gender):
        self.sid = sid
        self.name = name
        self.gender = gender


class Hostel:
    def __init__(self):
        self.rooms = {}

    def allocate_room(self, room_no, student):
        if room_no in self.rooms:
            print("Room already occupied!")
        else:
            self.rooms[room_no] = student
            print(f"{student.name} allocated to Room {room_no}")

    def vacate_room(self, room_no):
        if room_no in self.rooms:
            print(f"Room {room_no} vacated.")
            del self.rooms[room_no]
        else:
            print("Room already empty.")

    def view_status(self):
        print("\nRoom Occupancy Status")
        for room, student in self.rooms.items():
            print(f"Room {room} -> {student.name}")

    def allocation_report(self):
        print("\nAllocation Report")
        print("ID\tName\tGender\tRoom")

        for room, student in self.rooms.items():
            print(
                f"{student.sid}\t{student.name}\t"
                f"{student.gender}\t{room}"
            )

    def gender_report(self):
        print("\nBoys Rooms:")
        for room, student in self.rooms.items():
            if student.gender.lower() == "male":
                print(f"Room {room} -> {student.name}")

        print("\nGirls Rooms:")
        for room, student in self.rooms.items():
            if student.gender.lower() == "female":
                print(f"Room {room} -> {student.name}")



hostel = Hostel()

s1 = Student(1, "Arun", "Male")
s2 = Student(2, "Priya", "Female")
s3 = Student(3, "Rahul", "Male")

hostel.allocate_room(101, s1)
hostel.allocate_room(102, s2)
hostel.allocate_room(103, s3)

hostel.view_status()
hostel.allocation_report()
hostel.gender_report()

hostel.vacate_room(102)

hostel.view_status()
```
